### Laplacian edge detection

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
lap = cv2.Laplacian(gray, cv2.CV_64F)
lap = np.uint8(np.absolute(lap))

plt.imshow(lap, cmap='gray')
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
text_lap = str(pytesseract.image_to_string(lap, lang='ara'))

print(text_lap)
```

### Canny edge detection

```python
canny = cv2.Canny(gray, 150, 175)

plt.imshow(canny, cmap='gray')
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
text_canny = str(pytesseract.image_to_string(canny, lang='ara'))

print(text_canny)
```

### Sobel edge detection

```python
sobelx = cv2.Sobel(gray, cv2.CV_8U, 1, 0)
sobely = cv2.Sobel(gray, cv2.CV_8U, 0, 1)
sobel = cv2.bitwise_or(sobelx, sobely)

plt.imshow(sobel, cmap='gray')
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
text_sobel = str(pytesseract.image_to_string(sobel, lang='ara'))

print(text_sobel)
```

## Method 3: Segmenting Image + Pre-Processing

```python
#gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(gray, (7,7), 0)
thresh = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,11,30)

# Dilate to combine adjacent text contours
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (9,9))
dilate = cv2.dilate(thresh, kernel, iterations=4)

# Find contours, highlight text areas, and extract ROIs
cnts = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = cnts[0] if len(cnts) == 2 else cnts[1]
```

```python
plt.imshow(dilate, cmap='gray')
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
img2 = np.array(img)
```

```python
line_items_coordinates = []

for c in cnts:
    area = cv2.contourArea(c)
    x,y,w,h = cv2.boundingRect(c)
    image = cv2.rectangle(img2, (x,y), (x+w, y+h), color=(0,255,0), thickness=2)
    line_items_coordinates.append([(x,y), (x+w, y+h)])
```

```python
plt.imshow(img2, cmap='gray')
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

### Transforming Image to Text

```python
full_text = ''

for c in line_items_coordinates:
    cropped = img[c[0][1]:c[1][1], c[0][0]:c[1][0]] 
    gray_crop = cv2.cvtColor(cropped, cv2.COLOR_BGR2GRAY)
    ret, thresh_crop = cv2.threshold(gray_crop, 200, 255, cv2.THRESH_BINARY)
    text_cropped = str(pytesseract.image_to_string(thresh_crop, lang='ara', config=r'--oem 3 --psm 6'))
    print('\n-----------------------------------------------\n')
    print(text_cropped)
    full_text = full_text + '\n' + text_cropped
    
#print(full_text)
```

```python
gray_crop = cv2.cvtColor(cropped, cv2.COLOR_BGR2GRAY)
blur_crop = cv2.GaussianBlur(gray_crop,(1,1), cv2.BORDER_DEFAULT)

ret, thresh_crop = cv2.threshold(blur_crop, 120, 120, cv2.THRESH_BINARY)

sobelx_crop = cv2.Sobel(thresh_crop, cv2.CV_8U, 1, 0)
sobely_crop = cv2.Sobel(gray_crop, cv2.CV_8U, 0, 1)
sobel_crop = cv2.bitwise_or(sobelx_crop, sobely_crop)

plt.imshow(sobel_crop, cmap='gray')
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
```

```python
text_cropped = str(pytesseract.image_to_string(sobel_crop, lang='ara', config=r'--oem 3 --psm 6'))
print(text_cropped)
```
